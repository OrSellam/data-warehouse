## 2. Transform

In this part we prepare our data before loading it to the data warehouse. We removed duplicates rows, handled missing values, created a single formats for columns and more tanformation actions.

```python
def transform(df):
    df = df.copy()
    
    #Remove duplicate rows
    before = df.shape[0]
    df = df.drop_duplicates()
    after = df.shape[0]
    res = before - after
    print("Duplicate rows deleted:", res)
    
    
    #convert to datetime
    date_columns = [col for col in df.columns if any(kw in col.lower() for kw in ['date', 'timestamp', 'order_approved_at'])]
    for col in date_columns:
        df.loc[:, col] = pd.to_datetime(df[col], errors='coerce')
        print("Converted  to date", col)
    
    return df
        
    
    
    
    
```

### fill nulls in order_item_id 

```python
#find rows with nulls
nulls = order_items_copy['order_item_id'].isna()


for order_id in order_items_copy.loc[nulls, 'order_id'].unique():
    order = (order_items_copy['order_id'] == order_id)
    existing_ids = order_items_copy.loc[order & ~nulls, 'order_item_id'].dropna().astype(int).tolist()
    num_missing = order.sum() - len(existing_ids)
    
    #if the specific order has other rows with number in order_item_id
    if existing_ids:
        max_id = max(existing_ids + [0])
        full_range = set(range(1, max_id + num_missing + 2))  
        missing_ids = sorted(full_range - set(existing_ids))[:num_missing]
    #if this is the first product for the order
    else:
        missing_ids = list(range(1, num_missing + 1))
    
    order_items_copy.loc[order & nulls, 'order_item_id'] = missing_ids

#convert to int
order_items_copy['order_item_id'] = order_items_copy['order_item_id'].astype(int)
    
```

```python
multi_item_orders = (
    order_items_copy
    .groupby('order_id')
    .size()
    .reset_index(name='num_items')
    .query('num_items > 1')
)


orders_with_multiple_items = order_items_copy[order_items_copy['order_id'].isin(multi_item_orders['order_id'])]

orders_with_multiple_items = orders_with_multiple_items.sort_values(by=['order_id', 'order_item_id'])


print(orders_with_multiple_items.head(40))
```

### fill the only null in price column, we saw the next row (58) hold the same freight value so we assume it's the same price

```python
#fill null in the only row where is missing based on the freight_value of the second product in the order who contains the same price
missing_index = order_items_copy[order_items_copy['price'].isna()].index[0]
order_items_copy.at[missing_index, 'price'] = 85.00
order_items_copy.loc[57]
```

### fill nulls in product_id column based on the price and frieght freight_value 

```python
#find nulls in product_id
mask_missing = order_items_copy['product_id'].isna()

for idx, row in order_items_copy[mask_missing].iterrows():
    order_id = row['order_id']
    price = row['price']
    freight_value = row['freight_value']
    
    #find other rows in the same order who are holding a product id and have the same freight_value and product_id
    match = order_items_copy[
        (order_items_copy['order_id'] == order_id) &
        (order_items_copy['price'] == price) &
        (order_items_copy['freight_value'] == freight_value) &
        (order_items_copy['product_id'].notna()) ]
    
    #fill the product id 
    if not match.empty:
        filled_value = match['product_id'].iloc[0]
        order_items_copy.at[idx, 'product_id'] = filled_value
        print(f" row {idx} filled with product_id: {filled_value}")
```

```python
def clean_columns(df):

    if 'customer_city' in df.columns:
        # Capitalize customer city names
        df['customer_city'] = df['customer_city'].str.title()

    # Replace '@' with space and trim whitespace
        df['customer_city'] = df['customer_city'].str.replace('@', ' ', regex=False).str.strip()

    # Convert state names to uppercase
        df['customer_state'] = df['customer_state'].str.upper()

    # Fill missing 'region' based on customer_state from known values
        state_to_region = df[df['region'].notna()] \
        .drop_duplicates(subset=['customer_state', 'region']) \
        .set_index('customer_state')['region'].to_dict()

        df.loc[df['region'].isna(), 'region'] = \
        df.loc[df['region'].isna(), 'customer_state'].map(state_to_region)
        
    elif 'seller_city' in df.columns:
        # Capitalize customer city names
        df['seller_city'] = df['seller_city'].str.title()

    # Replace '@' with space and trim whitespace
        df['seller_city'] = df['seller_city'].str.replace('@', ' ', regex=False).str.strip()

    # Convert state names to uppercase
        df['seller_state'] = df['seller_state'].str.upper()

    # Fill missing 'region' based on customer_state from known values
        state_to_region = df[df['region'].notna()] \
        .drop_duplicates(subset=['seller_state', 'region']) \
        .set_index('seller_state')['region'].to_dict()

        df.loc[df['region'].isna(), 'region'] = \
        df.loc[df['region'].isna(), 'seller_state'].map(state_to_region)
    

    # Standardize region names
    region_mapping = {
        'Southeast': 'South East',
        'South': 'South',
        'Other': 'Other',
        'Central-West': 'Central West',
        'Northeast': 'North East'
    }

    df['region'] = df['region'].replace(region_mapping)

    return df
```

### remove string accents and change to title 

```python
import unicodedata

def remove_accents(text):
    return ''.join(
        c for c in unicodedata.normalize('NFKD', text)
        if not unicodedata.combining(c)
    )

geolocation_copy['geolocation_city'] = (
    geolocation_copy['geolocation_city'].str.strip().str.title().apply(remove_accents))

print(geolocation_copy['geolocation_city'].value_counts().head(10))
```

## create the schema and the tables in MySQL

```python
with engine.begin() as con:
    con.execute(text('DROP SCHEMA IF EXISTS olist'))
    con.execute(text('CREATE SCHEMA IF NOT EXISTS olist'))
    
    #use database drinks
    con.execute(text('USE olist'))
```

```python
with engine.begin() as con:
    con.execute(text("""
    CREATE TABLE IF NOT EXISTS olist.fact_order_items (
        order_id VARCHAR(50),
        order_item_id INT,
        product_id VARCHAR(50),
        seller_id VARCHAR(50),
        price FLOAT,
        freight_value FLOAT,
        customer_unique_id VARCHAR(50),
        order_purchase_timestamp DATETIME,
        order_date DATE,
        PRIMARY KEY (order_id, order_item_id)
    )
    """))
order_items_copy.to_sql('fact_order_items', con=engine, index=False, if_exists='append', schema='olist')
```

```python
with engine.begin() as con:
    con.execute(text("""
    CREATE TABLE IF NOT EXISTS olist.dim_order (
        order_id VARCHAR(50) PRIMARY KEY,
        customer_id VARCHAR(50),
        order_status VARCHAR(50),
        order_purchase_timestamp DATETIME,
        order_approved_at DATETIME,
        order_delivered_carrier_date DATETIME,
        order_delivered_customer_date DATETIME,
        order_estimated_delivery_date DATETIME,
        review_score INT,
        payment_type VARCHAR(50),
        full_or_installments VARCHAR(50),
        payment_installments INT)"""))
orders_copy.to_sql('dim_order', con=engine, index=False, if_exists='append', schema='olist')
```

```python
with engine.begin() as con:
    con.execute(text("""
    CREATE TABLE olist.dim_product (
        product_id VARCHAR(100) PRIMARY KEY,
        product_category_name VARCHAR(100),
        product_category_name_english VARCHAR(100),
        product_description TEXT,
        launch_date DATE)"""))
products_copy.to_sql('dim_product', con=engine, index=False, if_exists='append', schema='olist')
```

```python
with engine.begin() as con:
    con.execute(text("""
    CREATE TABLE IF NOT EXISTS olist.dim_seller (
        seller_id VARCHAR(50) PRIMARY KEY,
        seller_city VARCHAR(100),
        seller_state VARCHAR(10),
        seller_company_name VARCHAR(255),
        seller_email VARCHAR(255),
        region VARCHAR(50),
        registration_date DATE)"""))
    
sellers_copy.to_sql('dim_seller', con=engine, index=False, if_exists='append', schema='olist')

```
